```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

Graficas de B vs I 

```python
#corrientes:
I_r = pd.read_excel('magnetico.xlsx', usecols='B', skiprows=0)
I_c = pd.read_excel('magnetico.xlsx','espiras', usecols='B', skiprows=0)
I_s = pd.read_excel('magnetico.xlsx','selenoide', usecols='B', skiprows=0)

I_r = np.array(I_r)
I_c = np.array(I_c)
I_s = np.array(I_s)
```

```python
#campos magnetico:
B_r = pd.read_excel('magnetico.xlsx', usecols='G', skiprows=0)
B_c = pd.read_excel('magnetico.xlsx','espiras', usecols='G', skiprows=0)
B_s = pd.read_excel('magnetico.xlsx','selenoide', usecols='G', skiprows=0)

B_r = np.array(B_r)
B_c = np.array(B_c)
B_s = np.array(B_s)
```

GRAFICAS:

1. LINEA RECTA

```python
#Arreglo lineal.

#pendiente de la recta:

n_r= 10
r1 = sum(B_r*I_r)
r2 = np.mean(I_r)
r3 = np.mean(B_r)
r4 = sum(I_r**2)
r5= r2**2

m_r = (r1-(n_r*r2*r3))/(r4-n_r*r5)

b_r = (r3*r4-r2*r1)/(n_r*r4-r5)

def lineal(m,x,b):
  return m*x+b

y_r= lineal(m_r,I_r,b_r)  

print('la pendiente es:', m_r, 'punto de corte con el eje es:', b_r )
```

```python
fig, r = plt.subplots()
r.plot(I_r,B_r,color = 'tab:green',marker = '*')
r.plot(I_r,y_r,'g-.')
r.set_title('Campo en una geometria recta', loc = "center", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'g'})
r.set_xlabel('Corriente A')
r.set_ylabel('Campo magnetico mT')
r.legend(['campo magnetico', 'y=0.03687879x-5.03959683e-05 '])
plt.show()
```

2. CIRCULAR

```python
#Arreglo lineal.

#pendiente de la recta:

n_c= 10
c1 = sum(B_c*I_c)
c2 = np.mean(I_c)
c3 = np.mean(B_c)
c4 = sum(I_c**2)
c5= c2**2

m_c = (c1-(n_c*c2*c3))/(c4-n_c*c5)

b_c = (c3*c4-c2*c1)/(n_c*c4-c5)

def lineal(m,x,b):
  return m*x+b

y_c= lineal(m_c,I_c,b_c)  

print('la pendiente es:', m_c, 'punto de corte con el eje es:', b_c )
```

```python
fig, c = plt.subplots()
c.plot(I_c,B_c,color = 'tab:blue',marker = '*')
c.plot(I_c,y_c,'c-.')
c.set_title('Campo en una geometria circular', loc = "center", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'c'})
c.set_xlabel('Corriente A')
c.set_ylabel('Campo magnetico mT')
c.legend(['campo magnetico', 'y=0.02689394x-0.00034917 '])
plt.show()
```
